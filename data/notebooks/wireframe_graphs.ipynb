```python
# Data Read
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

sns.set_theme()
plt.figure(figsize=(6, 12))
dataset = pd.DataFrame(data={'Mon':[1],'Tue':[4],'Wed':[2],'Thur':[1],'Fri':[9],'Sat':[4],'Sun':[0]})

sns.barplot(data=dataset)
plt.ylabel("Interactions")
plt.xlabel("Day of the week")
plt.title("Interactions by Day")
```
